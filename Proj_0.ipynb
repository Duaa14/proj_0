### المهمة الثالثة

قم بكتابة دالة تقوم بمقارنة الرقم المدخل (مثل ١٩) بالرقم ١٠ و تخبرك ما إذا كان أكبر من الرقم ١٠ أم لا

```python
def bigger_than_10(x):
    if(x<10):
        print(str(x)+ " is smaller than 10")
    else:
        print(str(x)+ " is bigger than 10")
    
bigger_than_10(19)
```
